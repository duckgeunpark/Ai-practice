```python
import pandas as pd

scores = pd.Series([85, 92, 78, 96, 88])
print(scores)
# 0    85
# 1    92
# 2    78
# 3    96
# 4    88

# 인덱스를 직접 지정할 수도 있습니다
scores = pd.Series([85, 92, 78], index=["철수", "영희", "민준"])
print(scores["영희"])  # 92
```

```python
# DataFrame — 2차원 데이터
data = {
    "이름":  ["철수", "영희", "민준"],
    "나이":  [25, 28, 23],
    "점수":  [85, 92, 78]
}

df = pd.DataFrame(data)
print(df)
#    이름  나이  점수
# 0  철수   25   85
# 1  영희   28   92
# 2  민준   23   78
```

```python
# 3-3. 데이터 기본 확인

df.head()       # 상위 5행 출력 (기본값)
df.tail(3)      # 하위 3행 출력
df.info()       # 열 이름, 데이터 타입, 결측값 여부
df.describe()   # 수치형 열의 기초 통계 (평균, 표준편차, 최솟값 등)
df.shape        # (행 수, 열 수) 확인
```

```python
# 4-1. 열(column) 선택하기

# 열 하나 선택 → Series 반환
df["이름"]

# 열 여러 개 선택 → DataFrame 반환
df[["이름", "직급"]]

# 열 목록 확인
df.columns
```

```python
#  행(row) 선택 — loc vs iloc

# loc — 인덱스 이름으로 선택
df.loc[0]           # 인덱스가 0인 행
df.loc[0:2]         # 인덱스 0~2행 (끝 포함)
df.loc[0, "평가등급"]   # 0행의 점수 열

# iloc — 순서(정수)로 선택
df.iloc[0]          # 첫 번째 행
df.iloc[0:2]        # 0~1행 (끝 미포함, Python 슬라이싱 방식)
df.iloc[0, 2]       # 0행 2번째 열
```

```python
# 4-3. 조건으로 필터링하기

# 연봉이 6000 이상인 행만
df[df["연봉(만원)"] >= 6000]

# 부서가 "개발팀"이고 연봉이 5000 이상인 행
df[(df["부서"] == "개발팀") & (df["연봉(만원)"] >= 5000)]

# 이름이 "김민준"인 행
df[df["이름"] == "김민준"]
```

```python
# 4-4. 정렬하기
df.sort_values("연봉(만원)", ascending=False)  # 연봉 내림차순
df.sort_values(["부서", "연봉(만원)"])          # 부서 → 연봉 순으로 정렬
```

```python
# 5-1. 결측값(NaN) 처리

df.isnull()           # 각 셀의 결측 여부 (True/False)
df.isnull().sum()     # 열별 결측값 개수

# 결측값 제거
df.dropna()           # 결측값 있는 행 전체 삭제
df.dropna(axis=1)     # 결측값 있는 열 전체 삭제

# 결측값 채우기
df.fillna(0)               # 모든 결측값을 0으로
df["연봉(만원)"].fillna(df["연봉(만원)"].mean())  # 평균값으로 채우기
```

```python
# 5-2. 중복 데이터 제거
df.duplicated()       # 중복 행 확인
df.drop_duplicates()  # 중복 행 제거
```

```python
# 5-3. 데이터 타입 변환
df.dtypes                                      # 각 열의 데이터 타입 확인
df["연봉(만원)"] = df["연봉(만원)"].astype(float)  # 정수 → 실수
df["사번"] = df["사번"].astype(str)                # 문자 확인
```

```python
# 5-4. 열 조작

# 열 이름 변경
df.rename(columns={"연봉(만원)": "salary", "이름": "name"}, inplace=True)

# 새 열 추가
df["고연봉"] = df["salary"] >= 6000

# 열 삭제
df.drop(columns=["고연봉"], inplace=True)
```

```python
# 6-1. 기본 통계
df["salary"].sum()    # 합계
df["salary"].mean()   # 평균
df["salary"].count()  # 개수 (결측값 제외)
df["salary"].max()    # 최댓값
df["salary"].min()    # 최솟값
```

```python
# 6-2. 그룹별 분석 — groupby

# 부서별 평균 연봉
df.groupby("부서")["salary"].mean()

# 부서별 여러 통계 한 번에
df.groupby("부서")["salary"].agg(["mean", "max", "count"])
```

```python
# 6-3. 피벗 테이블

df.pivot_table(values="salary", index="부서", columns="직급", aggfunc="mean")
```
